```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
sys.path.append(os.getcwd()[:-8]+'/Analysis_Funcs/General')
from fitsloader import XMMtolc
from astropy.io import fits
```

Obs 1 full band

```python
lc1 = XMMtolc('obs/0604860201_.2-12.0_pn.fits')
lc1_start = lc1.time[0]
lc1 = lc1.rebin(250)
lc1 = lc1.shift(-lc1.time[0])
lc1.plot()
```

Obs 2 Full band

```python
lc2 = XMMtolc('obs/0604860301_.2-12.0_t50_pn.lc',t_bin=50)
lc2_start = lc2.time[0]
lc2 = lc2.rebin(250)
lc2 = lc2.shift(-lc2.time[0])
lc2.plot()
```

Obs 2 full band raw and background

```python
lc2_raw = XMMtolc('obs/0604860301/0604860301_.2-12.0_t50_raw.lc',t_bin=50).rebin(250)
lc2_bkg = XMMtolc('obs/0604860301/0604860301_.2-12.0_t50_bkg.lc',t_bin=50).rebin(250)
lc2_raw = lc2_raw.shift(-lc2_raw.time[0])
lc2_bkg = lc2_bkg.shift(-lc2_bkg.time[0])
plt.plot(lc2_raw.time,lc2_raw.countrate,color='b')
plt.plot(lc2_bkg.time,lc2_bkg.countrate,color='r')
plt.show()
```

...and zoom in on flare near end

```python
plt.plot(lc2_raw.time[300:],lc2_raw.countrate[300:],color='b')
plt.plot(lc2_bkg.time[300:],lc2_bkg.countrate[300:],color='r')
plt.show()
```

Obs 2 hard band vs soft band

```python
lc2_soft = XMMtolc('obs/0604860301_.2-2.0_t50_pn.lc',t_bin=50).rebin(250)
lc2_hard = XMMtolc('obs/0604860301_2.0-12.0_t50_pn.lc',t_bin=50).rebin(250)
lc2_soft = lc2_soft.shift(-lc2_soft.time[0])
lc2_hard = lc2_hard.shift(-lc2_hard.time[0])
plt.plot(lc2_hard.time,lc2_hard.countrate,color='b')
plt.plot(lc2_soft.time,lc2_soft.countrate,color='r')
plt.show()
```

...and zoom in on flare near end

```python
plt.plot(lc2_hard.time[300:],lc2_hard.countrate[300:],color='b')
plt.plot(lc2_soft.time[300:],lc2_soft.countrate[300:],color='r')
plt.show()
```

```python
full_lc = XMMtolc('obs/0604860201_.2-12.0_pn.fits').rebin(250).join(XMMtolc('obs/0604860301_.2-12.0_t50_pn.lc',t_bin=50).rebin(250)).join(XMMtolc('obs/0604860401_.2-12.0_t50_pn.lc',t_bin=50).rebin(250))
full_lc = full_lc.shift(-full_lc.time[0])
plt.plot(full_lc.time,full_lc.countrate)
plt.show()
```

Plot positions of two flares in obs 2 and 3 and track back into obs 1.

```python
lc2_relstart = lc2_start - lc1_start
lc3_relstart = lc3_start - lc1_start
lc2_peak = lc2_relstart + 113817.99530908857
lc3_peak = lc3_relstart + 23144.614621838566
t_per = lc3_peak - lc2_peak
print(t_per)
peak_positions = np.arange(lc2_peak-3*t_per,lc3_peak + 2*t_per,t_per)

plt.plot(full_lc.time,full_lc.countrate)
plt.vlines(peak_positions,0,0.25,colors='r',alpha=0.5)
```
